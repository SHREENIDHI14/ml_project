```python
import pandas as pd
import numpy as np
```

```python
text=pd.read_excel("dataset.xlsx")
text=pd.DataFrame(text)
text
```

```python
text=text.drop('inline-comment-id',axis=1)
text
```

```python
text['Question']=text['Question'].apply(lambda x:" ".join(x.lower() for x in x.split()))
text
```

```python
text['Question'] = text['Question'].str.replace('[^\w\s]','')
text
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
text['Question'] = text['Question'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
text
```

```python
frequency = pd.Series(' '.join(text['Question']).split()).value_counts()[:10]
frequency = list(frequency.index)
text['Question'] = text['Question'].apply(lambda x: " ".join(x for x in x.split() if x not in frequency))
text
```

```python
frequency = pd.Series(' '.join(text['Question']).split()).value_counts()[-10:]
frequency = list(frequency.index)
text['Question'] = text['Question'].apply(lambda x: " ".join(x for x in x.split() if x not in frequency))
text
```

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
text['Question']=text['Question'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
text["Final Label"]=text["Final Label"].replace({"suggestion":1,
           "hypothetical scenario":2,
           "rhetorical question":3,
           })
text.loc[text['Final Label'].str.contains('request',na=False), 'Final Label'] =4
text["Final Label"]=text["Final Label"].apply(lambda x: 5 if type(x)==str else x)
text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
X = count_vector.fit_transform(text["Question"]).toarray()

from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
X = tfidf_transformer.fit_transform(X)
Y = text['Final Label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.70)
```

```python
from sklearn import svm
S=svm.SVC()
S.fit(x_test.toarray(),y_test)
y_pred=S.predict(x_test.toarray())
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
